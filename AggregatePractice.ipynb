## Aggregate Query Practice

```python
import warnings
warnings.filterwarnings('ignore')
from reframe import Relation
country = Relation('/home/faculty/millbr02/pub/country.csv')
```

### Which government form has the highest average life expectancy?



```python
country.groupby("governmentform").mean("lifeexpectancy").sort("mean_lifeexpectancy", ascending = False).head(1)
```

### Calculate the surface area of the regions and sort in ascending order



```python
country.groupby("region").sum("surfacearea").rename("sum_surfacearea", "Surface Area").sort("Surface Area", ascending = True)
```

### which government types have the lowest average life expectancy?



```python
country.groupby("governmentform").mean("lifeexpectancy").sort("mean_lifeexpectancy")
```

### What regions have a high average life expectancy?



```python
country.groupby("region").mean("lifeexpectancy").sort("mean_lifeexpectancy", ascending = False)
```

### What is the most common form of government?



```python
country.groupby("governmentform").count("name").rename("count_name", "Countries").\
sort("Countries", ascending = False).head(1)
```

### Which region has the highest total gnp?



```python
country.groupby('region').sum('gnp').rename('sum_gnp', 'GNP').sort('GNP', ascending = False).head(1)
```

### Show the 10 government forms with the largest population



```python
country.groupby('governmentform').sum('population').rename('sum_population', 'Population').sort('Population', ascending = False).head(10)
```

### List the total population of each continent and order them from smallest to largest



```python
#country.groupby('continent').sum('population').rename('sum_population', 'Population').sort('Population')
country.groupby('region').sum('population').rename('sum_population', 'Population').sort('Population', ascending = False).head(10)
```

### Show the total number of countries for each region in Europe

```python
country.query("continent == 'Europe'").groupby('region').count('name').rename('count_name', 'Countries')
```

### List the total population of each region in Africa from lowest to highest.

```python
country.query("continent == 'Asia'").groupby('region').sum('population').rename('sum_population', 'Population').sort('Population', ascending = False)
```
